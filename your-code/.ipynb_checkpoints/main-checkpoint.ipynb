#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
month_steps = fitbit.groupby(by= ['Months_encoded'])['Steps'].median()

fig = plt.figure(figsize = (12,4))
plt.bar(month_steps.index, month_steps.values, color='#27756e')
plt.xticks(month_steps.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.savefig('figures/month_steps.png', dpi=600)
                 
# MEDIAN STEPS BY WORK_OR_WEEKEND
week_steps = fitbit.groupby(by= ['Work_or_Weekend'])['Steps'].median()

fig = plt.figure(figsize = (5,5))
plt.bar(week_steps.index, week_steps.values, color='#27756e')
plt.xticks([0,1],['Weekend', 'Work'])
plt.savefig('figures/week_steps.png', dpi=600)


```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
loop = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']

for minutes in loop:
    plt.figure()
    plt.title(minutes + 'VS Steps')
    plt.xlabel('Steps')
    plt.ylabel(minutes)
    plt.scatter(fitbit['Steps'], fitbit[minutes], color= np.random.rand(4))
```
